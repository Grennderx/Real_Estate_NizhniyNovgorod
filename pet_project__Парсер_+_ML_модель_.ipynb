```python
import pandas as pd
import numpy as np
```

```python
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv('nn_cian_2.csv')
df
```

```python
df = df.drop(df[df['Общая площадь'] == 'unknown'].index)
df
```

```python
df = df.set_axis([i for i in range(df.shape[0])], axis=0)
df
```

```python
df = df.drop('Дом', axis=1)
```

```python
df['Отделка'] = df['Отделка'].map({'unknown': 0, 'Неизвестно': 0, 'Предчистовая': 1, 'Без отделки': 2, 'Черновая': 3, 'С отделкой': 4, 'Чистовая': 5})
```

```python
df['Общая площадь'] = df['Общая площадь'].astype(float)
df['Этаж'] = df['Этаж'].astype(int)
```

```python
df.info()
```

```python
df[df['Площадь кухни'] == 'unknown']
```

```python
df[df['Жилая площадь'] == 'unknown']
```

```python
df_to_predict_square = df[df['Жилая площадь'] != 'unknown']
df_to_predict_square = df_to_predict_square[df_to_predict_square['Площадь кухни'] != 'unknown']
df_to_predict_square = df_to_predict_square[['Общая площадь', 'Жилая площадь', 'Площадь кухни', 'prices']]
df_to_predict_square['Жилая площадь'] = df_to_predict_square['Жилая площадь'].astype(float)
df_to_predict_square['Площадь кухни'] = df_to_predict_square['Площадь кухни'].astype(float)
df_to_predict_square
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_validate
```

```python
X = df_to_predict_square[['Общая площадь', 'prices']]
y1 = df_to_predict_square['Жилая площадь']
y2 = df_to_predict_square['Площадь кухни']
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y1, train_size = 0.75)

model = LinearRegression()

model.fit(Xtrain, ytrain)

y_living_pred = model.predict(Xtest)

print(r2_score(ytest, y_living_pred))
```

```python
df_predict = df[df['Жилая площадь'] == 'unknown']
fg = df_predict[['Общая площадь', 'prices']]
y_pred = model.predict(fg)
df_predict['Жилая площадь'] = y_pred
df[df['Жилая площадь'] == 'unknown'] = df_predict
(df['Жилая площадь'] == 'unknown').sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y2, train_size = 0.75)

from sklearn.ensemble import RandomForestRegressor

model2 = RandomForestRegressor()
model2.fit(X_train, y_train)

y_kitchen_pred = model2.predict(X_test)
print(r2_score(y_test, y_kitchen_pred))
```

```python
df_predict_kitchen = df[df['Площадь кухни'] == 'unknown']
X = df_predict_kitchen[['Общая площадь', 'prices']]
y_pred_kitchen = model.predict(X)
df_predict_kitchen['Площадь кухни'] = y_pred_kitchen
df[df['Площадь кухни'] == 'unknown'] = df_predict_kitchen
(df['Площадь кухни'] == 'unknown').sum()
```

```python
import random
a = list(set(df['Год'].values))
a.remove('unknown')
df.loc[df['Год'] == 'unknown', 'Год'] = random.choice(a)
```

```python
y = df['prices']
X = df.drop('prices', axis=1)

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, train_size = 0.75, random_state=42)
reg = RandomForestRegressor()
reg.fit(Xtrain, ytrain)

y_price_pred = reg.predict(Xtest)
print(r2_score(ytest, y_price_pred))
```

```python
# функция для предсказания цены
def predict_the_price(square, living_square, kitchen_square, floor, year, trim):
  feats = np.array([square, living_square, kitchen_square, floor, year, trim]).reshape(1, -1)
  return (f'{round(reg.predict(feats)[0])} рублей')
```
